## Question 1:
- Create  Numpy arrays for, x,y, where x and y are equal and both of them are between [0,11], and z = x^2 

### 1a) Create the first plot by creating a figure object located at [0,0,1,1] and plot (x,y) on this location and set the labels of your choice and make the title " x- y relation".  (as shown below!)




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

#state values for x,y,and z

x=np.linspace(0,10,11)
y=x
z=x**2

#create data for graphing

```

```python
fig = plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x, y, 'r')
axes.set_xlabel("Width")
axes.set_ylabel("Height")
axes.set_title("x-y relation")
```

## Question 5: 


- Create a bar chart that shows the "Average price in each state". ( The states should be represented on the X axis and the prices should be presented on the Y axis.  (as shown below)


```python
import usaddress   #parse us address
```

```python
addresslist = df["Address"]
addressdictlist = []
for address in addresslist:
  addressdict = {}
  parsed = usaddress.parse(address)
  for value, key in parsed:
    value = value.strip(",")
    if addressdict.get(key,"") == "":
      addressdict[key] = value
    else:
      addressdict[key] = addressdict[key] + " " + value
  addressdictlist.append(addressdict)
    
adf = pd.DataFrame.from_dict(addressdictlist)
adf
adf['Index'] = range(1, len(df) + 1)
adf
df = pd.concat([df, adf], axis=1)
df

# this creates a new data fram and joins it using the index to the original data frame 
#allows me to pull the state out of the address
```

```python
gdf=df[['StateName','Price']]
gdf=gdf.groupby(['StateName'], as_index=False).mean()
gdf

# creates grouping removing the index keeps the column headers equal
```

```python
plt.figure(figsize=(20, 5))
gdf.StateName = gdf['StateName'].astype(str)
plt.bar(x=gdf['StateName'], height=gdf['Price'])
plt.ticklabel_format(style='plain', axis='y')

#final graph with state and average. Error in the State names sometimes. Should have done the parse twice. Derived state name not from address but zip code
```
