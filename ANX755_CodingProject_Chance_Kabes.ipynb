```python
import pandas as pd
import numpy as np

x2015 = pd.read_csv("C:\\Users\\chanc\\Documents\\Creighton\\Data Wrangling\\2015.csv")
x2016 = pd.read_csv("C:\\Users\\chanc\\Documents\\Creighton\\Data Wrangling\\2016.csv")
x2017 = pd.read_csv("C:\\Users\\chanc\\Documents\\Creighton\\Data Wrangling\\2017.csv")
x2018 = pd.read_csv("C:\\Users\\chanc\\Documents\\Creighton\\Data Wrangling\\2018.csv")
x2019 =  pd.read_csv("C:\\Users\\chanc\\Documents\\Creighton\\Data Wrangling\\2019.csv")
```

```python
#add in year to each dataframe
x2015["Year"]=2015
x2016["Year"]=2016
x2017["Year"]=2017
x2018["Year"]=2018
x2019["Year"]=2019
```

```python
#clean 2015 with correct headers for append 
x2015.drop(columns=["Region","Standard Error","Dystopia Residual"], inplace = True)
#clean column headers with rename
x2015.rename(
    columns={"Country": "Country or region", "Happiness Rank": "Overall rank", "Happiness Score": "Score", "Economy (GDP per Capita)":"GDP per capita", "Family":"Social support","Health (Life Expectancy)":"Healthy life expectancy","Trust (Government Corruption)":"Perceptions of corruption"}
    ,inplace=True)
```

```python
#clean 2016 with correct headers for append 
x2016.drop(columns=["Region","Dystopia Residual","Lower Confidence Interval", "Upper Confidence Interval"], inplace = True)
#clean column headers with rename
x2016.rename(
    columns={"Country": "Country or region", "Happiness Rank": "Overall rank", "Happiness Score": "Score", "Economy (GDP per Capita)":"GDP per capita", "Family":"Social support","Health (Life Expectancy)":"Healthy life expectancy","Trust (Government Corruption)":"Perceptions of corruption"}
    ,inplace=True)
```

```python
#clean 2017 with correct headers for append 
x2017.drop(columns=["Dystopia.Residual","Whisker.high", "Whisker.low"], inplace = True)
#clean column headers with rename
x2017.rename(
    columns={"Country": "Country or region", "Happiness.Rank": "Overall rank", "Happiness.Score": "Score", "Economy..GDP.per.Capita.":"GDP per capita", "Family":"Social support","Health..Life.Expectancy.":"Healthy life expectancy","Trust..Government.Corruption.":"Perceptions of corruption"}
    ,inplace=True)
```

```python
#clean column headers with rename
x2018.rename(
    columns={"Freedom to make life choices": "Freedom"}
    ,inplace=True)
```

```python
x2019.rename(
    columns={"Freedom to make life choices": "Freedom"}
    ,inplace=True)
```

```python
world_happiness = pd.concat([x2015,x2016,x2017,x2018,x2019], axis =0) #concatenate all the values ontop of each other. Append required me to sort headers which I didn't want to do
world_happiness.info() #there is one value in 2018 that is null
```

## Q6

```python
def Happy_Indicator(x):
    Score = x[2]
    if Score <= 6:
        return "not-happy"
    else:
        return "happy"
```

```python
world_happiness["Happy-notHappy"] = world_happiness.apply(Happy_Indicator, axis = "columns")
world_happiness
```
